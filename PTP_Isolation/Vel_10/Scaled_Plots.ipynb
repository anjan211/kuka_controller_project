```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

Fetching all the trace files and storing them in two variables. One of them include trace files of Simulation for PTP with isolated axis at 10% velocity, while the other two variables include trace files of Real robot.

```python
# Fetch all files
PATH = './'
fileNames = os.listdir(PATH)
fileNames_Sim = [file for file in fileNames if '.csv' and 'KukaOL' in file]
fileNames_Real = [file for file in fileNames if '.csv' and 'WINDOWS' in file]
```

Scaling and Plotting the axis position of Real robot and Simulation for PTP with isolated axis at 10% velocity

```python
#Scaling of PTP with isolated axis at 10% velocity for real & simulation

li_Sim = []
li_Real =[]
if (len(fileNames_Sim)==len(fileNames_Real)):
    for i in range (len(fileNames_Sim)):
        df_Sim = pd.read_csv(fileNames_Sim[i], delimiter=';', encoding='cp1252')
        df_Real = pd.read_csv(fileNames_Real[i], delimiter=';', encoding='cp1252')
        li_Sim.append(df_Sim)
        li_Real.append(df_Real)

        # Plot Scaled Axis 1 position 
        if (i==0):
            time_Shift = li_Real[i]['Zeit[sec]'] - 0
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act1[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act1[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[7])
            plt.title('Simulation and Real position plots of A1 10% velocity')
            plt.savefig('Images/Scaled_Plots/A1.jpg')

        # Plot Scaled Axis 2 position 
        elif (i==1):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.012        #Shift Real robot trace by 0.012 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act2[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act2[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[8])
            plt.title('Simulation and Real position plots of A2 10% velocity')
            plt.savefig('Images/Scaled_Plots/A2.jpg')
        
        # Plot Scaled Axis 3 position 
        elif (i==2):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.052        #Shift Real robot trace by 0.052 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act3[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act3[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[9])
            plt.title('Simulation and Real position plots of A3 10% velocity')
            plt.savefig('Images/Scaled_Plots/A3.jpg')
        
        # Plot Scaled Axis 4 position 
        elif (i==3):
            time_Shift = li_Real[i]['Zeit[sec]'] - 0.027        #Shift Real robot trace by 0.027 towards left
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act4[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act4[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[10])
            plt.title('Simulation and Real position plots of A4 10% velocity')
            plt.savefig('Images/Scaled_Plots/A4.jpg')
        
        # Plot Scaled Axis 5 position 
        elif (i==4):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.012        #Shift Real robot trace by 0.012 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act5[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act5[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[11])
            plt.title('Simulation and Real position plots of A5 10% velocity')
            plt.savefig('Images/Scaled_Plots/A5.jpg')

        # Plot Scaled Axis 6 position 
        elif (i==5):
            time_Shift = li_Real[i]['Zeit[sec]'] - 0.12        #Shift Real robot trace by 0.12 towards left
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act6[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act6[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[12])
            plt.title('Simulation and Real position plots of A6 10% velocity')
            plt.savefig('Images/Scaled_Plots/A6.jpg')
            
```
