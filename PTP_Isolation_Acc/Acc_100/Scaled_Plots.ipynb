```python
import os
from os import path
import shutil
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

Fetching all the trace files and storing them in two variables. One of them include trace files of Simulation for PTP with isolated axis at 100% acceleration, while the other two variables include trace files of Real robot.

```python
# Fetch all files
target_path_1 = '../../Traces_Simulation/PTP_Acc_Sim'
target_path_2 = '../../Traces_Real/PTP_Acceleration_Real'
root_path = './'
fileNames_1 = os.listdir(target_path_1)
fileNames_2 = os.listdir(target_path_2)

fileNames_Sim = [file for file in fileNames_1 if '100_po85_KRCIpo.csv' in file and 'KukaOL' in file]
fileNames_Real = [file for file in fileNames_2 if '100_po85_KRCIpo.csv' in file and 'WINDOWS' in file]

for file in fileNames_Sim:
    shutil.copy(path.join(target_path_1, file), root_path)

for file in fileNames_Real:
    shutil.copy(path.join(target_path_2, file), root_path)
```

Scaling and Plotting the axis position of Real robot and Simulation for PTP with isolated axis at 100% acceleration

```python
#Scaling of PTP with isolated axis at 10% acceleration for real & simulation

li_Sim = []
li_Real =[]
if (len(fileNames_Sim)==len(fileNames_Real)):
    for i in range (len(fileNames_Sim)):
        df_Sim = pd.read_csv(fileNames_Sim[i], delimiter=';', encoding='cp1252')
        df_Real = pd.read_csv(fileNames_Real[i], delimiter=';', encoding='cp1252')
        li_Sim.append(df_Sim)
        li_Real.append(df_Real)

        # Plot Axis 1 position 
        if (i==0):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.022     #Shift Real robot trace by 0.022 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['AxisPos_Act1[]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['AxisPos_Act1[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[7])
            plt.title('Simulation and Real position plots of A1 100% acceleration')
            plt.savefig('Images/Plots/A1.jpg')
            plt.show()
            plt.close()


            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act1[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act1[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[7])
            plt.title('Simulation and Real scaled position plots of A1 100% acceleration')
            plt.savefig('Images/Scaled_Plots/A1.jpg')
            plt.show()
            plt.close()

        # Plot Axis 2 position 
        elif (i==1):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.018     #Shift Real robot trace by 0.018 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['AxisPos_Act2[]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['AxisPos_Act2[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[8])
            plt.title('Simulation and Real position plots of A2 100% acceleration')
            plt.savefig('Images/Plots/A2.jpg')
            plt.show()
            plt.close()

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act2[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act2[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[8])
            plt.title('Simulation and Real scaled position plots of A2 100% acceleration')
            plt.savefig('Images/Scaled_Plots/A2.jpg')
            plt.show()
            plt.close()
        
        # Plot Axis 3 position 
        elif (i==2):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.016        #Shift Real robot trace by 0.016 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['AxisPos_Act3[]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['AxisPos_Act3[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[9])
            plt.title('Simulation and Real position plots of A3 100% acceleration')
            plt.savefig('Images/Plots/A3.jpg')
            plt.show()
            plt.close()

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act3[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act3[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[9])
            plt.title('Simulation and Real scaled position plots of A3 100% acceleration')
            plt.savefig('Images/Scaled_Plots/A3.jpg')
            plt.show()
            plt.close()
        
        # Plot Axis 4 position 
        elif (i==3):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.027       #Shift Real robot trace by 0.027 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['AxisPos_Act4[]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['AxisPos_Act4[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[10])
            plt.title('Simulation and Real position plots of A4 100% acceleration')
            plt.savefig('Images/Plots/A4.jpg')
            plt.show()
            plt.close()

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act4[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act4[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[10])
            plt.title('Simulation and Real scaled position plots of A4 100% acceleration')
            plt.savefig('Images/Scaled_Plots/A4.jpg')
            plt.show()
            plt.close()
        
        # Plot Axis 5 position 
        elif (i==4):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.009       #Shift Real robot trace by 0.009 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['AxisPos_Act5[]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['AxisPos_Act5[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[11])
            plt.title('Simulation and Real position plots of A5 100% acceleration')
            plt.savefig('Images/Plots/A5.jpg')
            plt.show()
            plt.close()

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act5[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act5[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[11])
            plt.title('Simulation and Real scaled position plots of A5 100% acceleration')
            plt.savefig('Images/Scaled_Plots/A5.jpg')
            plt.show()
            plt.close()

        # Plot Axis 6 position 
        elif (i==5):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.03       #Shift Real robot trace by 0.03 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['AxisPos_Act6[]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['AxisPos_Act6[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[12])
            plt.title('Simulation and Real position plots of A6 100% acceleration')
            plt.savefig('Images/Plots/A6.jpg')
            plt.show()
            plt.close()

            #Scaling the data using 'MinMaxScalar'
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            #Labeling columns of the scaled data
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            
            #Plotting the scaled data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act6[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act6[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[12])
            plt.title('Simulation and Real scaled position plots of A6 100% acceleration')
            plt.savefig('Images/Scaled_Plots/A6.jpg')
            plt.show()
            plt.close()
            
```
