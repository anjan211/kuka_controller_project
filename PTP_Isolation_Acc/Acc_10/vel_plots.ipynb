```python
import os
from os import path
import shutil
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Fetch all files
target_path_1 = '../../Traces_Simulation/PTP_Acc_Sim'
target_path_2 = '../../Traces_Real/PTP_Acceleration_Real'
root_path = './'
fileNames_1 = os.listdir(target_path_1)
fileNames_2 = os.listdir(target_path_2)

fileNames_Sim = [file for file in fileNames_1 if 'A1_PTP_ACC_10_po85_NextGenDrive#1.csv' in file or 'A2_PTP_ACC_10_po85_NextGenDrive#2.csv' in file or 'A3_PTP_ACC_10_po85_NextGenDrive#3.csv' in file or 'A4_PTP_ACC_10_po85_NextGenDrive#4.csv' in file or 'A5_PTP_ACC_10_po85_NextGenDrive#5.csv' in file or 'A6_PTP_ACC_10_po85_NextGenDrive#6.csv' in file]
fileNames_Real = [file for file in fileNames_2 if 'A1_PTP_ACC_10_po85_NextGenDrive#1.csv' in file or 'A2_PTP_ACC_10_po85_NextGenDrive#2.csv' in file or 'A3_PTP_ACC_10_po85_NextGenDrive#3.csv' in file or 'A4_PTP_ACC_10_po85_NextGenDrive#4.csv' in file or 'A5_PTP_ACC_10_po85_NextGenDrive#5.csv' in file or 'A6_PTP_ACC_10_po85_NextGenDrive#6.csv' in file]

for file in fileNames_Sim:
    shutil.copy(path.join(target_path_1, file), root_path)

for file in fileNames_Real:
    shutil.copy(path.join(target_path_2, file), root_path)
```

```python
li_Sim = []
li_Real =[]
if (len(fileNames_Sim)==len(fileNames_Real)):
    for i in range (len(fileNames_Sim)):
        df_Sim = pd.read_csv(fileNames_Sim[i], delimiter=';', encoding='cp1252')
        df_Real = pd.read_csv(fileNames_Real[i], delimiter=';', encoding='cp1252')
        li_Sim.append(df_Sim)
        li_Real.append(df_Real)

        # Plot Axis 1 velocity 
        if (i==0):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.00    #Shift Real robot trace by 0.00 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['Istgeschwindigkeit[°/s]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['Istgeschwindigkeit[°/s]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[5])
            plt.title('Simulation and Real velocity plots of A1 10% acceleration')
            plt.savefig('Images/Velocity/A1.jpg')
            plt.show()
            plt.close()

        # Plot Axis 2 velocity 
        elif (i==1):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.00     #Shift Real robot trace by 0.0 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['Istgeschwindigkeit[°/s]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['Istgeschwindigkeit[°/s]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[5])
            plt.title('Simulation and Real velocity plots of A2 10% acceleration')
            plt.savefig('Images/Velocity/A2.jpg')
            plt.show()
            plt.close()

        # Plot Axis 3 velocity 
        elif (i==2):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.00     #Shift Real robot trace by 0.0 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['Istgeschwindigkeit[°/s]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['Istgeschwindigkeit[°/s]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[5])
            plt.title('Simulation and Real velocity plots of A3 10% acceleration')
            plt.savefig('Images/Velocity/A3.jpg')
            plt.show()
            plt.close()

        # Plot Axis 4 velocity 
        elif (i==3):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.00     #Shift Real robot trace by 0.0 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['Istgeschwindigkeit[°/s]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['Istgeschwindigkeit[°/s]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[5])
            plt.title('Simulation and Real velocity plots of A4 10% acceleration')
            plt.savefig('Images/Velocity/A4.jpg')
            plt.show()
            plt.close()

        # Plot Axis 5 velocity 
        elif (i==4):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.00     #Shift Real robot trace by 0.0 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['Istgeschwindigkeit[°/s]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['Istgeschwindigkeit[°/s]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[5])
            plt.title('Simulation and Real velocity plots of A5 10% acceleration')
            plt.savefig('Images/Velocity/A5.jpg')
            plt.show()
            plt.close()

        # Plot Axis 6 velocity 
        elif (i==5):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.00     #Shift Real robot trace by 0.0 towards right
            li_Real[i]['Zeit[sec]'] = time_Shift

            #Plotting the actual data
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(li_Sim[i]['Zeit[sec]'], li_Sim[i]['Istgeschwindigkeit[°/s]'], label = 'Simulation')
            plt.plot(li_Real[i]['Zeit[sec]'], li_Real[i]['Istgeschwindigkeit[°/s]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[5])
            plt.title('Simulation and Real velocity plots of A6 10% acceleration')
            plt.savefig('Images/Velocity/A6.jpg')
            plt.show()
            plt.close()

        
```
