```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Fetch all files
PATH = './'
fileNames = os.listdir(PATH)
fileNames_Sim1 = [file for file in fileNames if '.csv' and 'KRCIpo' in file]
fileNames_Real1 = [file for file in fileNames if '.csv' and 'KRCIpo' in file]

fileNames_Sim = [file for file in fileNames_Sim1 if 'KukaOL' in file]
fileNames_Real = [file for file in fileNames_Real1 if 'WINDOWS' in file]

fileNames_Away_Sim = [file for file in fileNames_Sim if '_away_10_' in file]
fileNames_Close_Sim = [file for file in fileNames_Sim if '_close_10_' in file]

fileNames_Away_Real = [file for file in fileNames_Real if '_away_10_' in file]
fileNames_Close_Real = [file for file in fileNames_Real if '_close_10_' in file]
```

```python
#Scaling of PTP Away real & simulation

li_Sim = []
li_Real =[]
if (len(fileNames_Away_Sim)==len(fileNames_Away_Real)):
    for i in range (len(fileNames_Away_Real)):
        df_Sim = pd.read_csv(fileNames_Away_Sim[i], delimiter=';', encoding='cp1252')
        df_Real = pd.read_csv(fileNames_Away_Real[i], delimiter=';', encoding='cp1252')
        li_Sim.append(df_Sim)
        li_Real.append(df_Real)

        if (i==0):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.055
            li_Real[i]['Zeit[sec]'] = time_Shift
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act1[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act1[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            # plt.xlim([0,1])
            # plt.ylim([0,1])
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[7])
            plt.title('Simulation and Real position plots of A1 100% velocity')
            if not os.path.exists('Images/Scaled_Plots/Away'):
                os.makedirs('Images/Scaled_Plots/Away')
            plt.savefig('Images/Scaled_Plots/Away/A1.jpg')

        elif (i==1):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.056
            li_Real[i]['Zeit[sec]'] = time_Shift
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act2[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act2[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            # plt.xlim([0,1])
            # plt.ylim([0,1])
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[8])
            plt.title('Simulation and Real position plots of A2 100% velocity')
            if not os.path.exists('Images/Scaled_Plots/Away'):
                os.makedirs('Images/Scaled_Plots/Away')
            plt.savefig('Images/Scaled_Plots/Away/A2.jpg')
        
        elif (i==2):
            time_Shift = li_Real[i]['Zeit[sec]'] + 0.02
            li_Real[i]['Zeit[sec]'] = time_Shift
            scalar = MinMaxScaler()
            scaled_Sim = scalar.fit_transform(li_Sim[i])
            scaled_Real = scalar.fit_transform(li_Real[i])
            scaled_Sim_df = pd.DataFrame(scaled_Sim, columns=li_Sim[i].columns)
            scaled_Real_df = pd.DataFrame(scaled_Real, columns=li_Real[i].columns)
            plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
            plt.plot(scaled_Sim_df['Zeit[sec]'], scaled_Sim_df['AxisPos_Act3[]'], label = 'Simulation')
            plt.plot(scaled_Real_df['Zeit[sec]'], scaled_Real_df['AxisPos_Act3[]'], label = 'Real robot')
            plt.legend()
            plt.grid()
            # plt.xlim([0,1])
            # plt.ylim([0,1])
            plt.xlabel(li_Real[i].columns[0])
            plt.ylabel(li_Real[i].columns[9])
            plt.title('Simulation and Real position plots of A3 100% velocity')
            if not os.path.exists('Images/Scaled_Plots/Away'):
                os.makedirs('Images/Scaled_Plots/Away')
            plt.savefig('Images/Scaled_Plots/Away/A3.jpg')
                   
```
