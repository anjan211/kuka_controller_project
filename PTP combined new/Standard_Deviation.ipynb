```python
import os
from os import path
import shutil
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from openpyxl import Workbook
import matplotlib.pyplot as plt
```

```python
wb = Workbook()
ws = wb.active
ws.cell(column=1, row=1, value='Combinations with A1')
ws.cell(column=1, row=3, value='Standard Deviation')
ws.cell(column=1, row=4, value='Maximum')
ws.cell(column=1, row=5, value='Mean')
```

```python
# Fetch all files
# Fetch all regular motion files
target_path_1 = '../Traces_Simulation/combined_ptp_new'
target_path_2 = '../Traces_Real/combined_new'
root_path = './'
fileNames_1 = os.listdir(target_path_1)
fileNames_2 = os.listdir(target_path_2)

fileNames_Sim_A1_2 = [file for file in fileNames_1 if 'A1_A2_2_NextGenDrive#1.csv' in file or 'A1_A3_2_NextGenDrive#1.csv' in file or 'A1_A4_2_NextGenDrive#1.csv' in file or 'A1_A5_2_NextGenDrive#1.csv' in file or 'A1_A6_2_NextGenDrive#1.csv' in file]
fileNames_Sim_A2_2 = [file for file in fileNames_1 if 'A2_A3_2_NextGenDrive#2.csv' in file or 'A2_A4_2_NextGenDrive#2.csv' in file or 'A2_A5_2_NextGenDrive#2.csv' in file or 'A2_A6_2_NextGenDrive#2.csv' in file]
fileNames_Sim_A3_2 = [file for file in fileNames_1 if 'A3_A4_2_NextGenDrive#3.csv' in file or 'A3_A5_2_NextGenDrive#3.csv' in file or 'A3_A6_2_NextGenDrive#3.csv' in file]
fileNames_Sim_A4_2 = [file for file in fileNames_1 if 'A4_A5_2_NextGenDrive#4.csv' in file or 'A4_A6_2_NextGenDrive#4.csv' in file]
fileNames_Sim_A5_2 = [file for file in fileNames_1 if 'A5_A6_2_NextGenDrive#5.csv' in file]

fileNames_Real_A1_2 = [file for file in fileNames_2 if 'A1_A2_2_NextGenDrive#1.csv' in file or 'A1_A3_2_NextGenDrive#1.csv' in file or 'A1_A4_2_NextGenDrive#1.csv' in file or 'A1_A5_2_NextGenDrive#1.csv' in file or 'A1_A6_2_NextGenDrive#1.csv' in file]
fileNames_Real_A2_2 = [file for file in fileNames_2 if 'A2_A3_2_NextGenDrive#2.csv' in file or 'A2_A4_2_NextGenDrive#2.csv' in file or 'A2_A5_2_NextGenDrive#2.csv' in file or 'A2_A6_2_NextGenDrive#2.csv' in file]
fileNames_Real_A3_2 = [file for file in fileNames_2 if 'A3_A4_2_NextGenDrive#3.csv' in file or 'A3_A5_2_NextGenDrive#3.csv' in file or 'A3_A6_2_NextGenDrive#3.csv' in file]
fileNames_Real_A4_2 = [file for file in fileNames_2 if 'A4_A5_2_NextGenDrive#4.csv' in file or 'A4_A6_2_NextGenDrive#4.csv' in file]
fileNames_Real_A5_2 = [file for file in fileNames_2 if 'A5_A6_2_NextGenDrive#5.csv' in file]

for file in fileNames_Sim_A1_2:
    shutil.copy(path.join(target_path_1, file), root_path)
for file in fileNames_Sim_A2_2:
    shutil.copy(path.join(target_path_1, file), root_path)
for file in fileNames_Sim_A3_2:
    shutil.copy(path.join(target_path_1, file), root_path)
for file in fileNames_Sim_A4_2:
    shutil.copy(path.join(target_path_1, file), root_path)
for file in fileNames_Sim_A5_2:
    shutil.copy(path.join(target_path_1, file), root_path)

for file in fileNames_Real_A1_2:
    shutil.copy(path.join(target_path_2, file), root_path)
for file in fileNames_Real_A2_2:
    shutil.copy(path.join(target_path_2, file), root_path)
for file in fileNames_Real_A3_2:
    shutil.copy(path.join(target_path_2, file), root_path)
for file in fileNames_Real_A4_2:
    shutil.copy(path.join(target_path_2, file), root_path)
for file in fileNames_Real_A5_2:
    shutil.copy(path.join(target_path_2, file), root_path)
```

```python
li_Sim_A1_2 =[]
li_Sim_A2_2 =[]
li_Sim_A3_2 =[]
li_Sim_A4_2 =[]
li_Sim_A5_2 =[]

li_Real_A1_2 =[]
li_Real_A2_2 =[]
li_Real_A3_2 =[]
li_Real_A4_2 =[]
li_Real_A5_2 =[]

if (len(fileNames_Sim_A1_2)==len(fileNames_Real_A1_2) and len(fileNames_Sim_A2_2)==len(fileNames_Real_A2_2) and len(fileNames_Sim_A3_2)==len(fileNames_Real_A3_2) and len(fileNames_Sim_A4_2)==len(fileNames_Real_A4_2) and len(fileNames_Sim_A5_2)==len(fileNames_Real_A5_2)):

    for i in range (len(fileNames_Sim_A1_2)):
        # Read csv
        df_Sim_A1_2 = pd.read_csv(fileNames_Sim_A1_2[i], delimiter=';', encoding='cp1252')
        df_Real_A1_2 = pd.read_csv(fileNames_Real_A1_2[i], delimiter=';', encoding='cp1252')
        li_Sim_A1_2.append(df_Sim_A1_2)
        li_Real_A1_2.append(df_Real_A1_2)

        # Scaling
        scalar = MinMaxScaler()
        scaled_Sim_A1_2 = scalar.fit_transform(li_Sim_A1_2[i])
        scaled_Real_A1_2 = scalar.fit_transform(li_Real_A1_2[i])
        scaled_Sim_df_A1_2 = pd.DataFrame(scaled_Sim_A1_2, columns=li_Sim_A1_2[i].columns)
        scaled_Real_df_A1_2 = pd.DataFrame(scaled_Real_A1_2, columns=li_Real_A1_2[i].columns)

        pos_col = df_Sim_A1_2.columns[5]
        diff = np.array([abs(scaled_Real_df_A1_2[pos_col] - scaled_Sim_df_A1_2[pos_col])])
        std = np.nanstd(diff)
        max = np.nanmax(diff)
        mean = np.nanmean(diff)
            
        # print (std)
        # print (max)
        # print (mean)

        # Write to excel
        ws.cell(column=i+2, row=1, value='A1-A'+str(i+2))
        ws.cell(column=i+2, row=3, value=std)
        ws.cell(column=i+2, row=4, value=max)
        ws.cell(column=i+2, row=5, value=mean)

    
    for i in range (len(fileNames_Sim_A2_2)):
        df_Sim_A2_2 = pd.read_csv(fileNames_Sim_A2_2[i], delimiter=';', encoding='cp1252')
        df_Real_A2_2 = pd.read_csv(fileNames_Real_A2_2[i], delimiter=';', encoding='cp1252')
        li_Sim_A2_2.append(df_Sim_A2_2)
        li_Real_A2_2.append(df_Real_A2_2)

        # Scaling
        scalar = MinMaxScaler()
        scaled_Sim_A2_2 = scalar.fit_transform(li_Sim_A2_2[i])
        scaled_Real_A2_2 = scalar.fit_transform(li_Real_A2_2[i])
        scaled_Sim_df_A2_2 = pd.DataFrame(scaled_Sim_A2_2, columns=li_Sim_A2_2[i].columns)
        scaled_Real_df_A2_2 = pd.DataFrame(scaled_Real_A2_2, columns=li_Real_A2_2[i].columns)

        plt.figure(figsize=(18,10), dpi=100, tight_layout=True)
        plt.plot(scaled_Sim_df_A2_2['Zeit[sec]'], scaled_Sim_df_A2_2['Istgeschwindigkeit[/s]'], label = 'Simulation')
        plt.plot(scaled_Real_df_A2_2['Zeit[sec]'], scaled_Real_df_A2_2['Istgeschwindigkeit[/s]'], label = 'Real')
        plt.legend()
        plt.grid()
        plt.title('PTP combined A2-A' + str(i+3) +' Real and Simulation velocity plots for A2')
        plt.savefig("Images/scaledA2_" + str(i) + ".jpg")
        plt.show()
        plt.close()

        pos_col = df_Sim_A2_2.columns[5]
        diff = np.array([abs(scaled_Real_df_A2_2[pos_col] - scaled_Sim_df_A2_2[pos_col])])
        std = np.nanstd(diff)
        max = np.nanmax(diff)
        mean = np.nanmean(diff)

        # Write to excel
        ws.cell(column=i+8, row=1, value='A2-A'+str(i+3))
        ws.cell(column=i+8, row=3, value=std)
        ws.cell(column=i+8, row=4, value=max)
        ws.cell(column=i+8, row=5, value=mean)


    for i in range (len(fileNames_Sim_A3_2)):
        df_Sim_A3_2 = pd.read_csv(fileNames_Sim_A3_2[i], delimiter=';', encoding='cp1252')
        df_Real_A3_2 = pd.read_csv(fileNames_Real_A3_2[i], delimiter=';', encoding='cp1252')
        li_Sim_A3_2.append(df_Sim_A3_2)
        li_Real_A3_2.append(df_Real_A3_2)

        # Scaling
        scalar = MinMaxScaler()
        scaled_Sim_A3_2 = scalar.fit_transform(li_Sim_A3_2[i])
        scaled_Real_A3_2 = scalar.fit_transform(li_Real_A3_2[i])
        scaled_Sim_df_A3_2 = pd.DataFrame(scaled_Sim_A3_2, columns=li_Sim_A3_2[i].columns)
        scaled_Real_df_A3_2 = pd.DataFrame(scaled_Real_A3_2, columns=li_Real_A3_2[i].columns)

        pos_col = df_Sim_A3_2.columns[5]
        diff = np.array([abs(scaled_Real_df_A3_2[pos_col] - scaled_Sim_df_A3_2[pos_col])])
        std = np.nanstd(diff)
        max = np.nanmax(diff)
        mean = np.nanmean(diff)

        # Write to excel
        ws.cell(column=i+13, row=1, value='A3-A'+str(i+4))
        ws.cell(column=i+13, row=3, value=std)
        ws.cell(column=i+13, row=4, value=max)
        ws.cell(column=i+13, row=5, value=mean)


    for i in range (len(fileNames_Sim_A4_2)):
        df_Sim_A4_2 = pd.read_csv(fileNames_Sim_A4_2[i], delimiter=';', encoding='cp1252')
        df_Real_A4_2 = pd.read_csv(fileNames_Real_A4_2[i], delimiter=';', encoding='cp1252')
        li_Sim_A4_2.append(df_Sim_A4_2)
        li_Real_A4_2.append(df_Real_A4_2)

       # Scaling
        scalar = MinMaxScaler()
        scaled_Sim_A4_2 = scalar.fit_transform(li_Sim_A4_2[i])
        scaled_Real_A4_2 = scalar.fit_transform(li_Real_A4_2[i])
        scaled_Sim_df_A4_2 = pd.DataFrame(scaled_Sim_A4_2, columns=li_Sim_A4_2[i].columns)
        scaled_Real_df_A4_2 = pd.DataFrame(scaled_Real_A4_2, columns=li_Real_A4_2[i].columns)

        pos_col = df_Sim_A4_2.columns[5]
        diff = np.array([abs(scaled_Real_df_A4_2[pos_col] - scaled_Sim_df_A4_2[pos_col])])
        std = np.nanstd(diff)
        max = np.nanmax(diff)
        mean = np.nanmean(diff)

        # Write to excel
        ws.cell(column=i+17, row=1, value='A4-A'+str(i+5))
        ws.cell(column=i+17, row=3, value=std)
        ws.cell(column=i+17, row=4, value=max)
        ws.cell(column=i+17, row=5, value=mean)


    for i in range (len(fileNames_Sim_A5_2)):
        df_Sim_A5_2 = pd.read_csv(fileNames_Sim_A5_2[i], delimiter=';', encoding='cp1252')
        df_Real_A5_2 = pd.read_csv(fileNames_Real_A5_2[i], delimiter=';', encoding='cp1252')
        li_Sim_A5_2.append(df_Sim_A5_2)
        li_Real_A5_2.append(df_Real_A5_2)

       # Scaling
        scalar = MinMaxScaler()
        scaled_Sim_A5_2 = scalar.fit_transform(li_Sim_A5_2[i])
        scaled_Real_A5_2 = scalar.fit_transform(li_Real_A5_2[i])
        scaled_Sim_df_A5_2 = pd.DataFrame(scaled_Sim_A5_2, columns=li_Sim_A5_2[i].columns)
        scaled_Real_df_A5_2 = pd.DataFrame(scaled_Real_A5_2, columns=li_Real_A5_2[i].columns)

        pos_col = df_Sim_A5_2.columns[5]
        diff = np.array([abs(scaled_Real_df_A5_2[pos_col] - scaled_Sim_df_A5_2[pos_col])])
        std = np.nanstd(diff)
        max = np.nanmax(diff)
        mean = np.nanmean(diff)

        # Write to excel
        ws.cell(column=i+20, row=1, value='A4-A'+str(i+6))
        ws.cell(column=i+20, row=3, value=std)
        ws.cell(column=i+20, row=4, value=max)
        ws.cell(column=i+20, row=5, value=mean)


# wb.save(filename='Standard_Deviation and Max_new.xlsx')

     
```
